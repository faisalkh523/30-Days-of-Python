```python
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}

# 1. Join A and B
A_join_B = A.union(B)
print(A_join_B)

# 2. Find A intersection B
A_intersection_B = A.intersection(B)
print(A_intersection_B)

# 3. Is A subset of B
A_subset_B = A.issubset(B)
print(A_subset_B)

# 4. Are A and B disjoint sets
A_disjoin_B = A.isdisjoint(B)
print(A_disjoin_B)

# 5. Join A with B and B with A
A_union_B = A.union(B)
B_union_A = B.union(A)
print(A_union_B)
print(B_union_A)

# 6. What is the symmetric difference between A and B
A_symmetric_difference_B = A.symmetric_difference(B)
print(A_symmetric_difference_B)

# 7. Delete the sets completely
del A,     # Delete set A
del B      # Delete set B
```

### 2. Explain the difference between the following data types: string, list, tuple and set
1. **string**: Text is a string data type. Any data type written as text is a string. Any data under single, double or triple quote are strings.
2. **list**: A list is collection of different data types which is ordered and modifiable(mutable). A list can be empty or it may have different data type items.
3. **tuple**: Tuple: is a collection which is ordered and unchangeable or unmodifiable(immutable). 
4. **set**: Set is a collection of items. Set is a collection of unordered and un-indexed distinct elements.

3. I am a teacher and I love to inspire and teach people. How many unique words have been used in the sentence? Use the split methods and set to get the unique words.

```python
the_sentence = 'I am a teacher and I love to inspire and teach people.'
split_sentence = the_sentence.split()           # split the sentence into words
print(split_sentence)

unique_words = set(split_sentence)      # removing duplicate words
print(unique_words)

amount_of_unique_words = len(unique_words)  # cpunt the unique words
print(amount_of_unique_words)
```
